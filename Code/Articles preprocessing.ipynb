```python
articles = pd.read_csv("../Data/export-articles.csv", encoding = 'utf8')
```

```python
complete_articles = articles[articles['text'].notnull()] # keep only the rows where we actually have the text
```

```python
complete_articles['externalid'] = complete_articles['externalid'].astype(str)
```

# Checking and correcting the articles with "anormal" text

```python
from bs4 import BeautifulSoup
```

```python
texts = list(complete_articles.text)
```

```python
texts_proper = []
for html in texts:
    soup = BeautifulSoup(html, features="html.parser")
    text = soup.get_text()
    texts_proper.append(text)
```

```python
complete_articles['text'] = texts_proper
```

```python
complete_articles.drop_duplicates('externalid').drop_duplicates('text').to_csv("../Data/articles_data.tsv", sep="\t", index = False)
```

```python
def save_texts():
    for _ , row in complete_articles.iterrows():
        index_ = row.externalid
        with open('../Data/textes_articles/' + str(index_) + ".txt", 'w', encoding = 'utf8') as f:
            f.write(row.text)
```
